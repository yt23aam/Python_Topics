<h3>Working with Dates and Time Series Data</h3>

Date Time Formatting Codes: [LINK](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

1.day_name() - returns the name of the day.

```python
import pandas as pd
```

```python
df = pd.read_csv('ETH_1h.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.loc[0, 'Date']
```

```python
try:
    df.loc[0, 'Date'].day_name()
except Exception as e:
    print(type(e))
    print(e)
```

```python
try:
    df['Date'] = pd.to_datetime(df['Date'])
except Exception as e:
    print(type(e))
    print(e)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %I-%p')
df['Date']
```

```python
df.loc[0, 'Date'].day_name()
```

```python
import datetime
d_parser = lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %I-%p')
df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
df.head()
```

```python
df.loc[0, 'Date'].day_name()
```

```python
df['Date'].dt.day_name()
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
df
```

```python
df['Date'].min()
```

```python
df['Date'].max()
```

```python
df['Date'].max() - df['Date'].min()
```

```python
filt = (df['Date'] >= '2020')
df.loc[filt]
```

```python
filt = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[filt]
```

```python
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[filt]
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df.loc['2019']
```

```python
df = df.sort_index()
df['2020-01': '2020-02'].sort_index(ascending=False)
```

```python
df['2020-01': '2020-02']['Close'].sort_index(ascending=False)
```

```python
df['2020-01': '2020-02']['Close'].mean()
```

```python
df.loc['2020-01-01', 'High'].max()
```

```python
df.loc['2020-01-01', 'High'].min()
```

```python
df['High'].resample('D').max()
```

```python
highs = df['High'].resample('D').max()
highs['2020-01-01']
```
