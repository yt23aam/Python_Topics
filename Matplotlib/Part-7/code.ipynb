<h5>Special Parameters</h5>

1. size parameter(s, float data-type)
2. color parameter(c, color as string)
3. marker parameter(marker, symbols)
4. border color for the point parameter(edgecolor, color as string)
5. linewidth of the marker edges(linewidth, float data-type)
6. opacity parameter(alpha, float data-type (0 to 1))

Click upon below Marker styles to check other marker styles

[Marker Styles](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.scatter(x, y, s=100, c='green', marker='.',
           edgecolor='black', linewidth=5, alpha=0.75)

plt.tight_layout()

plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
#          538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=100, c=colors, marker='.',
           edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()

plt.show()
```

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens', marker='.',
           edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()

plt.show()
```

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens', marker='.',
           edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()

plt.show()
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', marker='.',
           edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()

plt.show()
```

<h5>Data from a csv file</h5>

```python
data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,
           edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer',
           edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
